```python
#clean u tube text
import pandas as pd
import pdfplumber
import os
import glob

#CHANGE THIS ONLY
folder_name =["Amazon_Warehouse_Process_Assistant_Pros_and_Cons.pdf",
              "From_Day_1_To_Year_6_Working_At_Amazon.pdf",
              "Investigating_Why_So_Many_Amazon_Workers_Are_Dying.pdf",
              "Why_I_Quit_Working_At_An_Amazon_Warehouse_Job.pdf",
              "Why_So_Many_Amazon_Workers_Are_Getting_Hurt.pdf"
              ]

for pdf_path in glob.glob(f"{folder_name}"):
  if not os.path.exists(pdf_path):
      raise FileNotFoundError(f"{pdf_path} not found in current directory: {os.getcwd()}")

  # 1) Extract text safely
  text_parts = []
  with pdfplumber.open(pdf_path) as pdf:
      for page in pdf.pages:
          ptxt = page.extract_text()
          if ptxt:                # avoid None pages
              text_parts.append(ptxt)
  raw_text = "\n".join(text_parts).strip()

# 2) Put into a DataFrame and save raw CSV
df = pd.DataFrame({"filename": [os.path.basename(pdf_path)], "raw_text": [raw_text]})
df.to_csv("youtube_text.csv", index=False)   # raw saved

# 3) Apply your cleaning function to the raw_text column
df["cleaned_text"] = df["raw_text"].apply(clean_text)

# 4) Save final CSV with cleaned text
df.to_csv("full_cleaned.csv", index=False)

print("Saved youtube_text.csv (raw) and full_cleaned.csv (raw + cleaned).")
```

```python
import os

# Define the directory to search in
directory_path = '/content/'

# List all files and directories in the specified path
for filename in os.listdir(directory_path):
    # Check if the filename contains spaces and is not a directory itself
    if ' ' in filename and os.path.isfile(os.path.join(directory_path, filename)):
        old_filepath = os.path.join(directory_path, filename)
        new_filename = filename.replace(' ', '_')
        new_filepath = os.path.join(directory_path, new_filename)

        try:
            os.rename(old_filepath, new_filepath)
            print(f"Renamed: '{filename}' to '{new_filename}'")
        except Exception as e:
            print(f"Error renaming '{filename}': {e}")
    elif os.path.isdir(os.path.join(directory_path, filename)) and ' ' in filename:
        # Handle renaming directories if needed, though the prompt implies files
        old_dirpath = os.path.join(directory_path, filename)
        new_dirname = filename.replace(' ', '_')
        new_dirpath = os.path.join(directory_path, new_dirname)

        try:
            os.rename(old_dirpath, new_dirpath)
            print(f"Renamed directory:'{new_dirname}'")
        except Exception as e:
            print(f"Error renaming directory '{filename}': {e}")
```
