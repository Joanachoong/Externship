```python
import nltk
nltk.download('stopwords')
```

```python
import string
from nltk.corpus import stopwords

def clean_text(text):
    # Convert to lowercase and remove punctuation
    text = text.lower().translate(str.maketrans('', '', string.punctuation))

    # Split into words
    words = text.split()

    # Remove stopwords
    filtered = [word for word in words if word not in stopwords.words('english')]

    # Join back into a single string
    return ' '.join(filtered)

print(clean_text("The team ,is the best I’ve worked with"))

```

```python
#start cleaning the file

# call the file
import pandas as pd

from google.colab import drive
data = pd.read_csv('sample_dirty_text.csv')

# original data
print(data.head())

#clean the data by caliing the function
data['cleaned_text']=data['noisy_text'].apply(clean_text)

#check the result
print(data.head())

# import the data
data.to_csv('sample_cleaned_text.csv', index=False)
```
